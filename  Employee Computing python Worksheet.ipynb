```python
import matplotlib.pyplot as plt

# Data
modes = ['Cycle', 'Bus', 'Electric Bike', 'Walk/Run', 'Train', 'Motorbike/moped', 'Taxi']
emissions_per_mile = [0.049, 0.1644, 0.0148, 0.0932, 0.0571, 0.1829, 0.2392]

# Find highest, lowest, and moderate emission modes
highest_mode = modes[emissions_per_mile.index(max(emissions_per_mile))]
lowest_mode = modes[emissions_per_mile.index(min(emissions_per_mile))]
moderate_modes = [mode for mode, emission in zip(modes, emissions_per_mile) if 0.05 < emission < 0.2]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(modes, emissions_per_mile, color='red')

# Annotate highest, lowest, and moderate modes
plt.text(modes.index(highest_mode), max(emissions_per_mile), f'Highest: {highest_mode}', ha='center', va='bottom')
plt.text(modes.index(lowest_mode), min(emissions_per_mile), f'Lowest: {lowest_mode}', ha='center', va='bottom')

for mode in moderate_modes:
    plt.text(modes.index(mode), emissions_per_mile[modes.index(mode)], f'Moderate: {mode}', ha='center', va='bottom')

plt.title('Emissions per Mile by Mode of Transportation')
plt.xlabel('Mode of Transportation')
plt.ylabel('Emissions per Mile (Kg CO2e)')
plt.xticks(rotation=45)
plt.ylim(0, max(emissions_per_mile) + 0.05)

plt.tight_layout()
plt.show()
```
